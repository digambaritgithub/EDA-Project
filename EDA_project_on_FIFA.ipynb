```python
# check players salary.
player_salary = fifa[['short_name','wage_eur']]
```

```python
player_salary = player_salary.sort_values(by=['wage_eur'],ascending = False)
```

```python
player_salary.head()
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(player_salary['short_name'])[0:5],list(player_salary['wage_eur'])[0:5],color = ["blue","green","red","pink","orange"])
```

```python
plt.figure(figsize=(6,6))
sns.barplot(player_salary['short_name'][0:5],player_salary['wage_eur'][0:5])
plt.xticks(rotation=30)
plt.xlabel('Player')
plt.ylabel('wage_eur');
```

```python
# Defending
player_defend = fifa[['short_name','nationality','club','defending']]
player_defend.sort_values(by=['defending'],ascending=False).head()
```

```python
#youngest player
player_age = fifa[['short_name','age']]
player_age.sort_values(by=['age']).head()
```

```python
# best youngest players.
data_young = fifa[fifa['age']<=20].sort_values(by=['potential','value_eur'],ascending = [False,True])[0:5]
plt.figure(figsize=(6,6))
sns.barplot(y = 'short_name',x = 'wage_eur',data = data_young)
```

```python
#Which foot do footballers preffer?
fifa['preferred_foot'].value_counts()
```

```python
# plotting preferred foot
plt.figure(figsize = (6,6))
sns.barplot(fifa['preferred_foot'].value_counts().keys(),fifa['preferred_foot'].value_counts())
```

```python
#best player in the world.
data = fifa.sort_values(by=['overall'],ascending = False).head(10)
sns.barplot(x = 'age',y = 'short_name',data = data)
```
